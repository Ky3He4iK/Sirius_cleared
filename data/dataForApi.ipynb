# school info

```python
sch = pd.read_csv('schools_info.csv', sep=';', encoding='utf-8')
sch.head()
```

```python
sch = sch[['short_name', 'county', 'district_x', 'ekis_id', 'email', 'financing', 'first_class', 'legal_address', 'name', 'ogrn', 'okato', 'ou_classes', 'principal', 'site', 'extended_groups' ]]
```

```python
sch.drop_duplicates(inplace=True, subset=['short_name', 'county', 'district_x', 'ekis_id', 'email', 'financing', 'first_class', 'legal_address', 'name', 'ogrn', 'okato', 'ou_classes', 'principal', 'site', 'extended_groups' ])
```

```python
sch = sch.reset_index(drop=True)
```
