```python
#by_classes = marks.groupby(['class_right', 'subjectname']).mark.max().reset_index()
#Находим максимальные оценки по всем предметам у 11 классов
#max_mark = by_classes[by_classes['class_right'] == 11].iloc[:, 1:].set_index('subjectname').T.to_dict('int')['mark']
#max_mark
```

```python
max_mark = {'Английский устный': 20.0,
 'Английский язык': 100.0,
 'Биология': 100.0,
 'География': 100.0,
 'Информатика и ИКТ': 100.0,
 'Испанский устный': 20.0,
 'Испанский язык': 100.0,
 'История': 100.0,
 'Литература': 100.0,
 'Математика': 5.0,
 'Математика базовая': 5.0,
 'Математика профильная': 100.0,
 'Немецкий устный': 20.0,
 'Немецкий язык': 100.0,
 'Обществознание': 100.0,
 'Русский язык': 100.0,
 'Сочинение(Изложение)': 5.0,
 'Физика': 100.0,
 'Французcкий язык': 100.0,
 'Французский устный': 20.0,
 'Химия': 100.0}
```

```python
len(df_marks)
```

```python
#находим средние оценки по всем предметам
df_marks = df_marks.groupby(['ekis_id', 'year', 'class_right', 'subjectname']).mark.mean().reset_index()
df_marks = df_marks.rename(columns={'mark': 'mean_mark'})
```

```python
df_marks.head()
```

```python
len(df_marks)
```

```python
#убираем все оценки, равные NaN
df_marks = df_marks[df_marks.mean_mark.notnull()].reset_index(drop=True)
```

```python
len(df_marks)
```

```python
df_marks['st_mark'] = 0
```

```python
#Приводим оценки у 11х классов к 100бальной системе
for i in range(df_marks.shape[0]):
    if (df_marks.loc[i, 'class_right'] == 9):
        df_marks.loc[i, 'st_mark']=df_marks.loc[i, 'mean_mark']
    else:
        df_marks.loc[i, 'st_mark']=(df_marks.loc[i, 'mean_mark']*100)/max_mark[df_marks.loc[i, 'subjectname']]
```

```python
del df_marks['mean_mark']
```

```python
df_marks.head()
```

```python
len(df_marks)
```

```python
df_marks['ge_id'] = df_marks['year'].astype(str)+'.'+df_marks['class_right'].astype(str)+'.'+df_marks['subjectname'].astype(str)
```

```python
df_marks.head()
```

```python
df_marks = pd.pivot_table(df_marks, values='st_mark', index='ekis_id', columns='ge_id', aggfunc=np.sum, dropna=False).reset_index()
```

```python
df_marks.head()
```

```python
len(df_marks)
```

```python
df_marks.fillna(0, inplace=True)
```

```python
df_marks.head()
```

```python
df_marks.to_csv('exams.csv', sep=';', encoding='utf-8', index=False)
```
